# DATA LOADING

```python
imageData = np.loadtxt("MNISTnumImages5000_balanced.txt")
labelData = np.loadtxt("MNISTnumLabels5000_balanced.txt")
```

# 0 to 9 EXTRACTION + SHUFFLING

```python
train_images, train_labels, test_images, test_labels, val_images, val_labels = Preprocessing.dataSet(imageData, labelData, shuffle=True, val=0)
train_images = Preprocessing.min_max_normalization(train_images)
```

# MODEL

```python
model = SOFM.SOFM(12, 12, 784)
model.build("uniform", train_images) # "normal", "uniform", "PCA"
```

# LEARNING

```python
model.fit(train_images, eta_0=0.01 , num_epochs=10, batch_size=None, plot=True)
```

# ACTIVITY MAP PER LABEL

```python
train_images, train_labels, test_images, test_labels, val_images, val_labels = Preprocessing.dataSet(imageData, labelData, shuffle=False, val=0.0)
test_images = Preprocessing.min_max_normalization(test_images)

for label in range(10):

    records = test_images[label*100:(label+1)*100]
    model.plot_activation(records, f"Activity Map of label {label}")
```

# FEATURE MAP PLOT

```python
combined_matrix = np.zeros((12 * 28, 12 * 28))
for i in range(144):
    row = (i // 12) * 28
    col = (i % 12) * 28
    weight_image = np.transpose(model.weights_mat[i, :].reshape(28, 28))
    combined_matrix[row:row + 28, col:col + 28] = weight_image
fig, ax = plt.subplots()
ax.imshow(combined_matrix, cmap='gray')
plt.axis("off")
plt.show()
```
