# DATA LOADING

```python
imageData = np.loadtxt("MNISTnumImages5000_balanced.txt")
labelData = np.loadtxt("MNISTnumLabels5000_balanced.txt")
```

# 0 to 9 EXTRACTION + SHUFFLING




```python
train_images, train_labels, test_images, test_labels, val_images, val_labels = Preprocessing.dataSet(imageData, labelData, shuffle=True, val=0.3)
```

# MODEL

```python
model = Sequential.Sequential()

model.add(Input.Input(784))
l1 = Dense.Dense(50, "sigmoid")
l2 = Dense.Dense(10, "softmax")
model.add(l1)
model.add(l2)

model.compile(set_adam=True, eta=0.01) # eta =0.01 work well with Adam optimizer
```

# TRAINING

```python
accuracy, _, _, _ = Metrics.metrics(train_labels, [np.argmax(arr) for arr in model.predict(train_images)])
train_error_frac_before = 1-accuracy
accuracy, _, _, _ = Metrics.metrics(test_labels, [np.argmax(arr) for arr in model.predict(test_images)])
test_error_frac_before = 1-accuracy
```

```python
lr = 0.1 # (works well without Adam optimizer)
model.fit(train_images, train_labels, eta = lr, num_epochs = 20, batch_size = 300, loss_type = "MSE", plot=True, clip_grad = 10000000,
            validation_set = (val_images, val_labels))
```

# PLOTS

```python
accuracy, _, _, _ = Metrics.metrics(train_labels, [np.argmax(arr) for arr in model.predict(train_images)])
train_error_frac_after = 1-accuracy
accuracy, _, _, _ = Metrics.metrics(test_labels, [np.argmax(arr) for arr in model.predict(test_images)])
test_error_frac_after = 1-accuracy
```

## BAR PLOTS

```python
x1 = [train_error_frac_before, train_error_frac_after]
x2 = [test_error_frac_before, test_error_frac_after]

data = [x1, x2]
X = np.arange(2)
fig = plt.figure()
plt.bar(X + 0.00, data[0], color = 'r', width = 0.25)
plt.bar(X + 0.25, data[1], color = 'b', width = 0.25)
plt.ylabel('Error fraction')
plt.xticks([r+0.1 for r in range(len(x1))],
        ['before', 'after'])
plt.title('Error fraction before and after training')
plt.legend(labels=['train', 'test'])
plt.savefig("ErrorBarPlot.png")
```

## CONFUSION MATRIX

```python
train_pred = [np.argmax(arr) for arr in model.predict(train_images)]
test_pred = [np.argmax(arr) for arr in model.predict(test_images)]

Metrics.plot_confusion_matrix(train_pred, train_labels, "train_confusion")
Metrics.plot_confusion_matrix(test_pred, test_labels, "test_confusion")
```
